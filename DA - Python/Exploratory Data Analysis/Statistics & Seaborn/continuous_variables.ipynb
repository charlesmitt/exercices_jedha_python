```python
import pandas as pd
import seaborn as sns
import time

data = pd.read_csv("sales_predictions.csv")
```

```python
def check_date(d):
    try:  
        t = time.strptime(d,'%d.%m.%Y')
        return True
    except:
        return False

data["date_valide"] = data["date"].apply(lambda x : check_date(x))
```

```python
data_wrong_dates = data[data['date_valide']==False]
```
