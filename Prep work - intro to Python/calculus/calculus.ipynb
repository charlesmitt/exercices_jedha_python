```python
import sympy
```

```python
from sympy import *
from sympy.plotting import plot
```

```python
x, y = symbols('x, y', real = True)

expr = Eq(y, (x**3))

res = solve(expr, x)

p = plot(expr.rhs, (x, -10, 10), show = False, line_color = 'blue')

p.show()
```

```python
x = symbols('x', real = True)

expr = (x**2 + 2*x + 12)

res = diff(expr, x)

p = plot(expr, (x, -10, 10), show = False, line_color = 'blue')
p2= plot(res, (x, -10, 10), show = False, line_color = 'red')

p.extend(p2)
p.show()
```
