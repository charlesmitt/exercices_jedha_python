```python
import sqlite3
import pandas as pd

# Créer une connexion SQLite (base de données en mémoire ou fichier local)
conn = sqlite3.connect(":memory:")  # ":memory:" pour une base en mémoire

# Lire un fichier CSV avec pandas
df = pd.read_csv("financial_data.csv")

# Charger le dataframe dans SQLite
df.to_sql("google_sheets_data", conn, if_exists="replace", index=False)

# Exécuter une requête SQL sur la table
query = "SELECT * FROM google_sheets_data"
result = pd.read_sql_query(query, conn)

result
```
