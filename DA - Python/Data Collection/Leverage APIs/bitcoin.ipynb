```python
import time
dict_des_dix_cryptos = {}


# on parcourt chaque crypto selon le DF créé plus tôt
for i in range(len(df_ten)):
    
    # pour chaque crypto on crée un nouveau dictionnaire qui contiendra les données
    dict_crypto = {}

    # récupère et affiche l'ID de la crypto
    id=df_ten.iloc[i,0]
    print(id)

    url = f"https://api.coingecko.com/api/v3/coins/{id}/ohlc"
    headers = {"accept": "application/json"}
    p = {'vs_currency': 'usd', 'days':365}

    response = requests.get(url, headers=headers, params=p).json()
    
    # pour chaque ligne de la réponse ( = une date ), on crée une entrée dans le dict avec la date comme clé et les OHLC comme valeurs
    for ligne in response:
        liste = [datetime.datetime.fromtimestamp(e/1000).strftime("%Y-%m-%d") if isinstance(e, int) else e for e in ligne]
        dict_crypto[liste[0]]= liste[1:]
    time.sleep(5)

    # le dictionnaire de la crypto est ajouté au gros dictionnaire avec l'id comme clé
    dict_des_dix_cryptos[id]=dict_crypto
```
