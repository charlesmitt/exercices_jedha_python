```python
import os
import logging
import scrapy
from scrapy.crawler import CrawlerProcess

class FirstMovieSpider(scrapy.Spider):
    # Name of your spider
    name = "firstmovie"

    # Url to start your spider from
    start_urls = [
        'https://www.imdb.com/chart/top'
    ]

    def first_movie(self, response):
        print("first_movie function")
        quote = response.css('div.ipc-page-content-container')

        crew = response.css('li.ipc-metadata-list__item')

        return {
            'title': quote.css('span.hero__primary-text::text').get(),
            'note' : quote.css('span.sc-d541859f-1::text').get(),
            'votes': quote.css('div.sc-d541859f-3::text').get(),
            'plot' : quote.css('span.sc-3ac15c8d-2::text').get(),

            'director' : crew[0].css('a.ipc-metadata-list-item__list-content-item::text').getall(),
            'writer' : crew[1].css('a.ipc-metadata-list-item__list-content-item::text').getall(),
            'actors': crew[2].css('a.ipc-metadata-list-item__list-content-item::text').getall()
        }

    # Callback function that will be called when starting your spider
    def parse(self, response):
        try:
            first_movie = response.css('a.ipc-title-link-wrapper').attrib['href']
        except:
            logging.info("No next page")
        else:
            logging.info("SUCCESS > FIRST PAGE")
            yield response.follow(first_movie, callback=self.first_movie)

        

filename = 'first_imdb.json'

if filename in os.listdir("json/"):
    os.remove('json/'+filename)

process = CrawlerProcess(settings={
    'USER_AGENT': 'Chrome/97.0',
    'LOG_LEVEL': logging.INFO,
    'FEEDS':{
        'json/' + filename : {"format": "json"},
        },
    "AUTOTHROTTLE_ENABLED": True
    }
)

#Lancer le crawler utilisant la spider
process.crawl(FirstMovieSpider)
process.start()
```
