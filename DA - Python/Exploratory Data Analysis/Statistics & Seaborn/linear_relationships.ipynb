```python
import pandas as pd
import seaborn as sns

data = pd.read_csv("IBM_HR_ATTRITION(1).csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data["Attrition"] = data["Attrition"].map({"Yes":1, "No":0})
```

```python
data["Attrition"] = data["Attrition"].apply(lambda x : 1 if x=='Yes' else 0)
```

```python
data.head()
```

```python
sns.lmplot(data=data, y='will_leave', x = 'DistanceFromHome', logistic=True, scatter_kws={"alpha":.05})
```

```python
sns.lmplot(data=data, y='Attrition', x = 'YearsAtCompany', logistic=True, scatter_kws={"alpha":.05})
```

```python
sns.lmplot(data=data, y='Attrition', x = 'YearsWithCurrManager', logistic=True, scatter_kws={"alpha":.05})
```

```python
sns.lmplot(data=data, y='Attrition', x = 'Age', logistic=True, scatter_kws={"alpha":.05})
```
