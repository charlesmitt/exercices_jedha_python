```python
import pandas as pd
import csv
import datetime

print("Menu")
print("====")
print("1 - Enregistrer nouvelle date clé")
print("2 - Voir dates clés")
print("3 - Supprimer dates clés")
print("4 - Voir tous les progrès")
print("5 - Quitter")
print("====================")
choice = input("Choose an option [1-5]")

badges = pd.read_csv( "badges.csv")

data = pd.read_csv( "data.csv")

if choice == '1':
    nouveau = input("saisir l'évènement et la date au format [jj-mm-aaaa], et l'heure [hh:mm] séparés par des virgules")
    #ADD CHECK DU FORMAT DE LA LIGNE SAISIE
    #
    # RECUPERATION DE LA DATE/HEURE DE L'EVENEMENT ET TRANSFO EN DATETIME 
    list = nouveau.split(",")
    name_event = list[0]
    date_full = list[1].split("-")
    date_day = int(date_full[0])
    date_month = int(date_full[1])
    date_year = int(date_full[2])

    hour_full = list[2].split(":")
    hour_hour = int(hour_full[0])
    hour_min = int(hour_full[1])
    datetime_event = datetime.datetime(date_year, date_month, date_day, hour_hour, hour_min)
    with open('data.csv', mode='a', newline='') as fichier:
        writer = csv.writer(fichier)
        writer.writerow([name_event, datetime_event])
elif choice == '2':
    print(data)
elif choice =='3':
    print(data)
    input("choisir quel évènement supprimer [id]")
elif choice == '4':
    datetime_now = datetime.datetime.now()
    datetime_epoch = datetime.datetime(1970, 1, 1)
    with open('data.csv') as file_obj: 
        data_csv = csv.reader(file_obj) 
        next(data_csv, None)
        for row in data_csv:    
            try: 
                
                #AFFICHAGE NOM DE L'EVENEMENT
                str_titre =   row[0]+" - "+row[1]             
                for x in range(len(str_titre)+18):
                    print("#",end="")
                print("\n#        "+str_titre+"        #")
                for x in range(len(str_titre)+18):
                    print("#",end="")
                
                
                datetime_event = datetime.datetime.strptime(row[1],"%Y-%m-%d %H:%M:%S")
                timedelta_event_epoch = datetime_event-datetime_epoch
                timedelta_now_event = datetime_now-datetime_event
                seconds_now_event = int(timedelta_now_event.total_seconds())
                
                print("")
                print("  ********************")
                print("  *  Badges obtenus  *")
                print("  ********************")
                print("")
                
                with open('badges.csv') as file_obj: 
                    badges_csv = csv.reader(file_obj) 
                    next(badges_csv, None)
                
                    for i in badges_csv:
                        seconds_duree_badge = int(i[1])
                        seconds_difference = seconds_now_event-seconds_duree_badge
                        if seconds_difference>0:
                            temps_badge = int(timedelta_event_epoch.total_seconds())+int(i[1])
                            date_badge = datetime.datetime.fromtimestamp(temps_badge).strftime('%Y-%m-%d %H:%M')
                            print("     "+i[0]+", obtenu le "+ date_badge)
                        else:

                            print("")
                            print("  ********************")
                            print("  *  Prochain badge  *")
                            print("  ********************")
                            print("")

                            print("     "+str(i[0])+", dans "+str(round(-seconds_difference/60/60))+" heures")
                            
                            progress =seconds_now_event/seconds_duree_badge*100
                            percentage_sur_20 = (round(progress/100*20))
                            
                            print("     ", end="")
                            for x in range(percentage_sur_20):
                                print("=",end="")
                            for y in range(20-percentage_sur_20):
                                print("-", end="")
                            print("  "+str(round(progress))+"%")
                            break

                    print("")
            except TypeError as e:
                print("e"+e)
elif choice == '5':
    exit
```
