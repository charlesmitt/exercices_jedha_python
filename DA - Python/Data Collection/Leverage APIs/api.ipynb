```python
import requests
ad = requests.get("https://api-adresse.data.gouv.fr/search/?q=8+rue+Guy+de+Maupassant&limit=20")
```

```python
df= pd.json_normalize(ad.json().get('features'))
```

```python
df["long"]=df["geometry.coordinates"].apply(lambda x : x[0])
df["lat"]=df["geometry.coordinates"].apply(lambda x : x[1])
```

```python

df.head(15)
```

```python
import plotly.express as px
fig = px.scatter_mapbox(df,lat="lat", lon="long",
mapbox_style="open-street-map", zoom=6, color_continuous_scale="Reds", text="properties.label")
fig.show()
```
